```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.dpi'] = 300


# Data for the experiments
experiments = ['Simulation 1', 'Simulation 2']
A = [4028.168364521667, 2411.5884429762827]
B = [1728.69364106235, 1432.8165310247048]

# Width of the bars
bar_width = 0.35

# Indices for the bars
index = np.arange(len(experiments))

# Create the bar plot
fig, ax = plt.subplots()

# Plot bars for condition A
bars_A = ax.bar(index, A, bar_width, label='No D-list', color='blue', alpha=0.6)

# Plot bars for condition B
bars_B = ax.bar(index + bar_width, B, bar_width, label='With D-list', color='orange', alpha=0.6)


# Set labels and title
ax.set_ylabel('RMSE')
ax.set_title('RSEM simulation results: 15151 out of 60604 genes indexed')
ax.set_xticks(index + bar_width / 2)
ax.set_xticklabels(experiments)
ax.legend()

# Show the plot
plt.show()
```
